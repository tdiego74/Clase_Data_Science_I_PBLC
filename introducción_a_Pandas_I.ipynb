```python
modelos = ['A4 3.0 Quattro 4dr manual',
 'A4 3.0 Quattro 4dr auto',
 'A6 3.0 4dr',
 'A6 3.0 Quattro 4dr',
 'A4 3.0 convertible 2dr']
peso = [3583, 3627, 3561, 3880, 3814]
precios = ['$33,430', '$34,480', '$36,640', '$39,640', '$42,490']
```

```python
# otra forma de generar el dataframe es primero uniendo las series:

s = pd.concat([Autos_peso, Autos_precio], axis=1)
s
```

```python
Numeros = range(30, 50, 2)
Numeros_serie = pd.Series(Numeros)

Numeros_en_texto = ['primeroo','segundo','tercero','cuarto','quinto','sexto','séptimo','octavo','noveno','décimo']
Numeros_serie_3 = pd.Series(Numeros,index=Numeros_en_texto)
Numeros_serie_3
```

```python
"""
Cuando se suman dos series en Python utilizando el operador +, si una serie tiene un valor NaN (Not a Number) y
la otra serie tiene un valor numérico, el resultado de la suma será NaN.

Esto se debe a que NaN representa una falta de valor o un valor desconocido, y cualquier operación matemática que
involucre NaN dará como resultado NaN. Por lo tanto, cuando se suman dos números, si uno de ellos es NaN, el resultado de
la suma también será NaN.
"""
Numeros_nan=Numeros_serie_2 + Numeros_serie_3
Numeros_nan
```
